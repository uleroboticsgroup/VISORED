```python
from paz.models import SSD300
import cv2
import time
import numpy as np
from paz.pipelines import PreprocessImage, PreprocessBoxes
from paz.models.detection.utils import create_prior_boxes
import os
```

```python
model = SSD300(3, base_weights=None, head_weights=None)
model.load_weights('object_detection/trained_models/SSD300/weights.hdf5')
preprocess_image = PreprocessImage((300, 300))
```

```python
times = []
for f in os.listdir('voc_dataset/VOC2007/JPEGImages'):
    img = cv2.imread('voc_dataset/VOC2007/JPEGImages/'+f)
    init = time.time()
    img_pr = preprocess_image(img)
    img_pr = np.expand_dims(img_pr, axis=0)    
    inferences = model.predict(img_pr)
    end = time.time()
    times.append(end-init) 
```

```python
def reject_outliers(data, m = 2.):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
times_without_outliers = reject_outliers(np.array(times))
```

```python
print('Inference time(S): ', times_without_outliers.mean())
print('FPS: ', (1/times_without_outliers.mean()))
```
