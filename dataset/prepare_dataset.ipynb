```python
import os
import pandas as pd
import labelme
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
columns = ['id', 'url', 'location', 'latitude', 'longitude', 'wolves', 'detections']
```

```python
for folder in ['europe', 'world']:
    info = pd.read_csv('wolf_'+folder+'.csv')
    data = pd.DataFrame(columns = columns)
    for index, row in info.iterrows():
        name_file = str(int(row.url_id)) if not pd.isna(row.url_id) else ''
        if os.path.exists( folder+'/'+name_file+'.png'):
            if os.path.exists( folder+'/'+ name_file+'.json'):
                label_file = labelme.LabelFile(filename=folder+'/'+ name_file+'.json')                
                cont = 0
                detections = np.array([])
                for shape in label_file.shapes:
                    cont +=1
                    detections = np.append( detections, np.array(shape['points']).flatten())
                location = [float(numeric_string) for numeric_string in row.location.split(',')]
                info_row = {
                    'id': row.url_id,
                    'url': row.url,
                    'latitude': location[0], 
                    'longitude' :location[1],
                    'wolves': int(cont),
                    'detections': ' '.join(map(str, detections.astype(np.int32))),
                    'location' : folder,
                }
                data = data.append(info_row, ignore_index=True)
            else:
                os.remove( os.path.join(folder, name_file+'.png') )
    data.to_csv('dataset_'+folder+'.csv', index=False)
```
