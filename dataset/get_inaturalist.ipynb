```python
iteraction = 0
dataSize = 1
df = pd.DataFrame([])
while (dataSize > 0):     
    iteraction = iteraction + 1
    url = 'https://api.inaturalist.org/v1/observations?taxon_id='+str(taxon_id)+'&place_id='+placeID+'&term_id=22&term_value_id=24&geo=true&order=desc&order_by=created_at&page='+str(iteraction)+'&per_page=200'
    response = requests.get(url, headers=token)
    dictr = response.json()
    recs = dictr['results']
    dataSize = len(recs)
    data = pd.json_normalize(recs)
    if dataSize > 0:
        df = df.append(data)
df.to_csv('wolf.csv', index=False)
df.shape
```

```python
df_new = pd.DataFrame(columns= df.columns)
for index, row in df.iterrows():
    if len(row.photos)>1:
        for id_photo, photo in zip(range(len(row.photos)), row.photos):
            row_new = row.copy()
            row_new['url'] = photo['url'].replace("square", "original")
            row_new['url_id'] = photo['id']
            urllib.request.urlretrieve(row_new['url'], 'europe/'+str(int(row_new['url_id']))+'.png')
            row_new['hash'] = hashlib.md5(open('europe/'+str(int(row_new['url_id']))+'.png','rb').read()).hexdigest()
            df_new = df_new.append(row_new, ignore_index = True)
df_new.to_csv('wolf_europe.csv', index=False)
df_new.shape
```
